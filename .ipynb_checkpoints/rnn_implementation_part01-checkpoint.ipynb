# Recurrent neural networks

RNNは可変の長さのシークエンスを扱えるニューラルネットです。（feedforward netsとは異なる）一般に、下記の再帰関係式で表せます。

云々

RNNはfeedback loopのある状態モデルように見えます。
状態は再帰関係式によって時が経つごとに変わり、一タイムステップ遅れてフィードバックが返ってきます。
この遅れてくるフィードバックループがモデルに記憶を与えます、なぜなら、状態のタイムステップ間で情報を憶えているからです。

あるK時点でのネットワークの最終的なアウトプットは、その前の各タイムステップの状態に基づき計算されています。

私たちはある時点でのinputとその前の状態を元に、現在の状態を計算することもできますし、今の状態と今のinputを元に次の状態を予測することもできます。この表記の違いは私たちのモデルに大きな影響は与えず、手元でのタスクによります。

このチュートリアルでは、recurrent netsはそれほどfeedforward netsと違わないことを説明します、ですがそれを訓練させるには少し難しいところがあります、それについて説明していきましょう。

# Linear recurrent neural network

チュートリアルの最初のパートでは、シンプルなRNNについてです。binary input streamの中に、いくつ1があったかを数え、outputとしてシークエンスの最後にそのトータルの数を返します。

ここでのRNNモデルは一つの状態をもっていて、binary streamから各タイムステップごとにinputが1つあり、そしてシークエンスの最後にその状態をアウトプットします。
下記の図のようなモデルになります。
左の図を展開すると右のようなネットワークになります。
展開されたネットワークは、各層で同じ共有のパラメータwrecとwxを持った、N+1の層のニューラルネットになります。

私たちはこのタスクの最適なパラメータはwrec,wxともに1にすることであることは容易に分かります。ですが、訓練とこのニューラルネットの実装は興味深いでしょう。
はじめに、必要なライブラリをimportし、データセットを定義しましょう。




```python
# Python imports
import numpy as np # Matrix and vector computation package
import matplotlib
import matplotlib.pyplot as plt  # Plotting library
from matplotlib import cm  # Colormaps
from matplotlib.colors import LogNorm  # Log colormaps
# Allow matplotlib to plot inside this notebook
%matplotlib inline
# Set the seed of the numpy random number generator so that the tutorial is reproducable
np.random.seed(seed=1)
```

# Define the dataset
この例でのinputデータXは、10のtime stepsごとに20のバイナリsequenceで構成されています。input sequenceはそれぞれ、0か1に丸められた一定のランダム分布で作られています。
目標outputtは、sequence内で1があった回数とします。sequenceはバイナリなので、1の合計値と等しくなります。

```python
# Create dataset
nb_of_samples = 20
sequence_len = 10
# Create the sequences
# シークエンスを作成(20x10)。ランダムに0、1
X = np.zeros((nb_of_samples, sequence_len))
for row_idx in range(nb_of_samples):
    X[row_idx,:] = np.around(np.random.rand(sequence_len)).astype(int)
# Create the targets for each sequence
# それぞれのシークエンスについて目標値を設定(各層について)
t = np.sum(X, axis=1)
```

# Training with backpropagation through time

recurrent netsを訓練するのに典型的なアルゴリズムはbackpropagation through timeアルゴリズムです。その名前は前のfeedforward netsのチュートリアルでお話ししたbackpropagationアルゴリズムに基づいていることをほのめかしています。

もし標準的なbackpropagationを理解していたら、backpropagation through timeの理解はそれほど難しくはありません。違いとしては、ある量のtimestepの間、recurrent netが展開される必要がある点のみです。この展開はすでにチュートリアルの初めに図示しています。展開後、標準的な多層feedforward networkによく似たモデルを得ます。違いは、それぞれの層が複数のinput(前の状態と現在のinput)を得ることと、パラメーター（$w_{rec}$と$w_x$）がそれぞれの層で共有されていることだけです。

# Compute the output with the forward step

forward stepではネットワークを展開し、forwardの活性化を標準的なbackpropagationのように計算することです。最終的なアウトプットはコスト関数を計算するのに使われます。コスト関数からは訓練に用いられる誤差シグナルが導き出されます

recurrent netを複数のtimestepにわたり展開する際に、各層で異なるtimestepごとに同様のrecurrence relationを計算します。モデルにおけるこのrecurrence relationは状態更新(update_state)メソッドと定義されます。

forward-statesメソッドはupdate_stateメソッドをforループ内で当てはめることで増加するtimesteps中で状態を計算します。複数のsequenceの為のforward stepはベクトル化を用いて並行処理されることをご認識ください、これは前のfeedforward netsのチュートリアルで図示されています。networkは何のsequenceも得ることなく始まるので、最初の状態を与える必要があります。この例では初めの状態を0に設定します（他のパラメータを与えることも可能です。）

最後に、outputの段階で、コストξは平均二乗誤差法(mean squared error function(MSE))でinput dataのすべてのsequenceについて計算されます。

```python
# Define the forward step functions
def update_state(xk, sk, wx, wRec):
    """
    Compute state k from the previous state (sk) and current input (xk),
    by use of the input weights (wx) and recursive weights (wRec).
    
    前の状態skと現在のinput　xkを元に、
    inputの重みwxと再帰的な重みwRecを用いて状態kを計算します。
    """
    return xk * wx + sk * wRec

def forward_states(X, wx, wRec):
    """
    Unfold the network and compute all state activations given the input X,
    and input weights (wx) and recursive weights (wRec).
    Return the state activations in a matrix, the last column S[:,-1] contains the
    final activations.
    
    ネットワークを展開し、input X、inputの重み、再帰的な重みによる状態活性化を計算してください。
    状態活性化は行列の形で帰ってきます。Sの最終列には最後の活性化が含まれています。
    """
    # Initialise the matrix that holds all states for all input sequences.
    # The initial state s0 is set to 0.
    # 縦はXと同じ、横はXより1つ多いarrayにする。S0があるため？
    S = np.zeros((X.shape[0], X.shape[1]+1))
    # Use the recurrence relation defined by update_state to update the 
    #  states trough time.
    for k in range(0, X.shape[1]):
        # S[k] = S[k-1] * wRec + X[k] * wx
        # input Xと前の状態Sを元にSを計算し代入する
        S[:,k+1] = update_state(X[:,k], S[:,k], wx, wRec)
    return S

def cost(y, t): 
    """
    Return the MSE between the targets t and the outputs y.
    目標tとoutput yの平均二乗誤差を返します。
    """
    return ((t - y)**2).sum() / nb_of_samples
```

# Compute the gradients with the backward step

backward stepはnetworkのoutputに関して、コストの勾配dξ/d
yを、output_gradientメソッドを用いて計算することから始めます。この勾配はbackward_gradientメソッドを用いて、アウトプットからインプットに向かって、後ろから時間を通して（一層一層）、伝播しパラメータを更新していきます。networkを通して勾配が伝播していく再帰関係式は下記のように記せます

$$ dξ　$$

そして、下記で始まります

$$ dξ $$

timestep数nとともに、networkは展開されます。状態に接続している再帰パラメータ$w_{rec}$のみが誤差をネットワークに伝播する役割を果たしています。

パラメータに関するコスト関数の勾配は各層のパラメータ勾配を合計したものと見なせます（もしくは誤差を伝播する間、勾配を蓄積するものとして）。重みの更新規則は下記の通りです。

$$ d $$

```python
def output_gradient(y, t):
    """
    Compute the gradient of the MSE cost function with respect to the output y.
    output yに関して平均二乗誤差コストファンクションの勾配を計算する
    """
    return 2.0 * (y - t) / nb_of_samples

def backward_gradient(X, S, grad_out, wRec):
    """
    Backpropagate the gradient computed at the output (grad_out) through the network.
    Accumulate the parameter gradients for wX and wRec by for each layer by addition.
    Return the parameter gradients as a tuple, and the gradients at the output of each layer.
    
    ネットワークを通してでたoutputについて計算した勾配(grad_out)をバックプロパゲーションします。
    付け足していくことで、各レイヤーのwX、wRecのパラメーター勾配を蓄積します。
    タプルの形式でパラメーター勾配と、各層のアウトプットでの勾配を返します。
    """
    # Initialise the array that stores the gradients of the cost with respect to the states.
    grad_over_time = np.zeros((X.shape[0], X.shape[1]+1))
    grad_over_time[:,-1] = grad_out
    # Set the gradient accumulations to 0
    wx_grad = 0
    wRec_grad = 0
    for k in range(X.shape[1], 0, -1):
        # Compute the parameter gradients and accumulate the results.
        wx_grad += np.sum(grad_over_time[:,k] * X[:,k-1])
        wRec_grad += np.sum(grad_over_time[:,k] * S[:,k-1])
        # Compute the gradient at the output of the previous layer
        grad_over_time[:,k-1] = grad_over_time[:,k] * wRec
    return (wx_grad, wRec_grad), grad_over_time
```

# Gradient Checking
feedforward netsについてのチュートリアルのパート4でもやったように、勾配チェックをまた行います。勾配を計算している際にミスをしていない、と言い切りたいですからね。勾配チェックはバックプロパゲーションで計算した勾配と、数値勾配が近いことを保証します。

```python
# Perform gradient checking
# Set the weight parameters used during gradient checking
params = [1.2, 1.2]  # [wx, wRec]
# Set the small change to compute the numerical gradient
eps = 1e-7
# Compute the backprop gradients
S = forward_states(X, params[0], params[1])
grad_out = output_gradient(S[:,-1], t)
backprop_grads, grad_over_time = backward_gradient(X, S, grad_out, params[1])
# Compute the numerical gradient for each parameter in the layer
for p_idx, _ in enumerate(params):
    grad_backprop = backprop_grads[p_idx]
    # + eps
    params[p_idx] += eps
    plus_cost = cost(forward_states(X, params[0], params[1])[:,-1], t)
    # - eps
    params[p_idx] -= 2 * eps
    min_cost = cost(forward_states(X, params[0], params[1])[:,-1], t)
    # reset param value
    params[p_idx] += eps
    # calculate numerical gradient
    grad_num = (plus_cost - min_cost) / (2*eps)
    # Raise error if the numerical grade is not close to the backprop gradient
    if not np.isclose(grad_num, grad_backprop):
        raise ValueError('Numerical gradient of {:.6f} is not close to the backpropagation gradient of {:.6f}!'.format(float(grad_num), float(grad_backprop)))
print('No gradient errors found')
```

# Updating the Parameters
Recurrent netsは学習させるのが難しいことで悪名高い。それは不安定な勾配のせいで、最急降下法のような勾配に基づく単純な最適化手法で良い局所最小値を見つけることが難しくなるからだ。

不安定さは下記の図で表現される。上の2つの図はある値$w_x$と$w_{rec}$の間のコスト面の部分を示している。色のついたマーカーは私たちが選んだbackprpagating through time中の変化の場所に置いてあります。変化の過程は最下部の図に描かれています。Cost surfaceの色は対数表現であることに気をつけてください。Cost surfaceは$w_x = w_{rec} =1$のあたりの小さい谷で0に近い値をとります。そして図の淵あたりではとても早くコストが増加します（特に$|w_{rec}|>1$で）

私たちのシンプルな線形モデルの勾配の不安定さの原因は、最下部の図に示されています。不安定さは下記の事実によって引き起こされています。それは勾配を時を遡って電番させるためにrecurrence relationがgeometric sequenceを形成していることです。(上記した通り)

d

状態$S_k$と$m$timesteps前の$ S_{k-m}$の勾配は下記のように記せます。

w

私たちのシンプルな線形モデルでは、もし$|w_{rec}|>1$なら勾配は指数関数的に増加することに気をつけてください（勾配爆発として知られています）。そしてもし$|w_{rec}|<1$なら勾配は指数関数的に小さくなっていきます(勾配消失として知られています)。

勾配爆発は$w_x = 1$、$w_{rec} = 2$にサンプルされた点で表現されています。それだと勾配はbackpropagationの間指数関数的に増えていきます(y軸はlog尺度です)。$w_x = 1$、$w_{rec} = -2$の点は、上下に勾配が爆発していく異なる種類を示しています、それは$w_{rec}$がマイナスであることに起因します。この勾配爆発が意味するのは、学習が$w_{rec}$パラメータを変化させることに敏感であるということです。

勾配消失は$w_x = 1$、$w_{rec} = 0.5$と$w_x = 1$、$w_{rec} = -0.5$の点で示されています。ともに指数関数的に勾配が減っています。この勾配消失が意味するのは、勾配は時を経るごとに消えるため、networkが長期間の従属を学習するのは難しいということです。

もし$w_{rec}=0$なら勾配はすぐに0となり、$w_{rec}=1$なら勾配は時が経ってもある値にとどまることに気をつけてください。

次のセクションでは、この不安定なエラー関数をどのように最適化するかについて説明します。

```python
# Define plotting functions

# Define points to annotate (wx, wRec, color)
points = [(2,1,'r'), (1,2,'b'), (1,-2,'g'), (1,0,'c'), (1,0.5,'m'), (1,-0.5,'y')]

def get_cost_surface(w1_low, w1_high, w2_low, w2_high, nb_of_ws, cost_func):
    """Define a vector of weights for which we want to plot the cost."""
    w1 = np.linspace(w1_low, w1_high, num=nb_of_ws)  # Weight 1
    w2 = np.linspace(w2_low, w2_high, num=nb_of_ws)  # Weight 2
    ws1, ws2 = np.meshgrid(w1, w2)  # Generate grid
    cost_ws = np.zeros((nb_of_ws, nb_of_ws))  # Initialize cost matrix
    # Fill the cost matrix for each combination of weights
    for i in range(nb_of_ws):
        for j in range(nb_of_ws):
            cost_ws[i,j] = cost_func(ws1[i,j], ws2[i,j])
    return ws1, ws2, cost_ws

def plot_surface(ax, ws1, ws2, cost_ws):
    """Plot the cost in function of the weights."""
    surf = ax.contourf(ws1, ws2, cost_ws, levels=np.logspace(-0.2, 8, 30), cmap=cm.pink, norm=LogNorm())
    ax.set_xlabel('$w_{in}$', fontsize=15)
    ax.set_ylabel('$w_{rec}$', fontsize=15)
    return surf

def plot_points(ax, points):
    """Plot the annotation points on the given axis."""
    for wx, wRec, c in points:
        ax.plot(wx, wRec, c+'o', linewidth=2)

def get_cost_surface_figure(cost_func, points):
    """Plot the cost surfaces together with the annotated points."""
    # Plot figures
    fig = plt.figure(figsize=(10, 4))   
    # Plot overview of cost function
    ax_1 = fig.add_subplot(1,2,1)
    ws1_1, ws2_1, cost_ws_1 = get_cost_surface(-3, 3, -3, 3, 100, cost_func)
    surf_1 = plot_surface(ax_1, ws1_1, ws2_1, cost_ws_1 + 1)
    plot_points(ax_1, points)
    ax_1.set_xlim(-3, 3)
    ax_1.set_ylim(-3, 3)
    # Plot zoom of cost function
    ax_2 = fig.add_subplot(1,2,2)
    ws1_2, ws2_2, cost_ws_2 = get_cost_surface(0, 2, 0, 2, 100, cost_func)
    surf_2 = plot_surface(ax_2, ws1_2, ws2_2, cost_ws_2 + 1)
    plot_points(ax_2, points)
    ax_2.set_xlim(0, 2)
    ax_2.set_ylim(0, 2)
    # Show the colorbar
    fig.subplots_adjust(right=0.8)
    cax = fig.add_axes([0.85, 0.12, 0.03, 0.78])
    cbar = fig.colorbar(surf_1, ticks=np.logspace(0, 8, 9), cax=cax)
    cbar.ax.set_ylabel('$\\xi$', fontsize=15, rotation=0, labelpad=20)
    cbar.set_ticklabels(['{:.0e}'.format(i) for i in np.logspace(0, 8, 9)])
    fig.suptitle('Cost surface', fontsize=15)
    return fig

def plot_gradient_over_time(points, get_grad_over_time):
    """Plot the gradients of the annotated point and how the evolve over time."""
    fig = plt.figure(figsize=(6.5, 4))  
    ax = plt.subplot(111)
    # Plot points
    for wx, wRec, c in points:
        grad_over_time = get_grad_over_time(wx, wRec)
        x = np.arange(-grad_over_time.shape[1]+1, 1, 1)
        plt.plot(x, np.sum(grad_over_time, axis=0), c+'-', label='({0}, {1})'.format(wx, wRec), linewidth=1, markersize=8)
    plt.xlim(0, -grad_over_time.shape[1]+1)
    # Set up plot axis
    plt.xticks(x)
    plt.yscale('symlog')
    plt.yticks([10**8, 10**6, 10**4, 10**2, 0, -10**2, -10**4, -10**6, -10**8])
    plt.xlabel('timestep k', fontsize=12)
    plt.ylabel('$\\frac{\\partial \\xi}{\\partial S_{k}}$', fontsize=20, rotation=0)
    plt.grid()
    plt.title('Unstability of gradient in backward propagation.\n(backpropagate from left to right)')
    # Set legend
    leg = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False, numpoints=1)
    leg.set_title('$(w_x, w_{rec})$', prop={'size':15})
    
def get_grad_over_time(wx, wRec):
    """Helper func to only get the gradient over time from wx and wRec."""
    S = forward_states(X, wx, wRec)
    grad_out = output_gradient(S[:,-1], t).sum()
    _, grad_over_time = backward_gradient(X, S, grad_out, wRec)
    return grad_over_time
```

```python
# Plot cost surface and gradients

# Get and plot the cost surface figure with markers
fig = get_cost_surface_figure(lambda w1, w2: cost(forward_states(X, w1, w2)[:,-1] , t), points)

# Get the plots of the gradients changing by backpropagating.
plot_gradient_over_time(points, get_grad_over_time)
# Show figures
plt.show()
```

# Rprop optimisation

前のセクションでは勾配がとても不安定であることを説明しました。勾配が不安定なエリアに入ると、Cost surfaceにおいて大きくジャンプしました、そしてoptimizerはもとの点から大きく離れた点に着地しました。これは下記の図で表されます。(Pascanu et alによる)

前のfeedforward netsのチュートリアルで、勾配降下法の更新ルールは下記だったことを思い出してください。

$W(i+1) = W(i)-μ \frac{∂ξ}{∂W(i)}$

$W(i)$は勾配降下法での各iteration $i$における$W$で、$μ$ は学習レートです。

もし学習中に上の面の青い点($w_x = 1$、$w_{rec} = 2$)に着地してしまったら、勾配は$10^7$台になってしまいます。たとえ、学習レートを小さい値$0.00001(10^{-6})$にしても、パラメータ$W$は現時点から10ユニットの距離を更新するでしょう。それは私たちの例では壊滅的です。これに対処する1つの手はさらに学習レートを下げることですが、もし最適化が低い勾配のエリアに入ったら、それ以上更新を動かさないという手もあります。

研究者はこの不安定な環境における勾配に基づく学習を扱うたくさんの手法を見つけています。例えばこんなものがあります:Gradient clipping,Hessian-Free Optimization Momentumなどです。

私たちは、最適化の更新を勾配に対して鈍感にすることで、不安定な勾配を扱えるようになります。その1つのやり方はresilent backpropagation(Rprop)と呼ばれるテクニックを用いることです。Rpropは前のチュートリアルでも述べたmomentum methodに似ていますが、パラメータを更新する際に、勾配の符号しか用いません。Rpropアルゴリズムは下記のようにまとめられます。

* 重みの更新初期値Δはゼロ以外の値にする
* それぞれのパラメータ$w$について：
  * もし重みの勾配の符号が前のiterationと比べて変わっていたら次の通り：　　
  $sign(\partial \xi /\partial w(i)) \neq sign(\partial \xi /\partial w(i-1))$

    * 重みの更新値 $\Delta$ には $\eta^-$ をかけてください、なお $\eta^- < 1$ です。  
    $\Delta(i) = \Delta(i) * \eta^-$   
  * もし勾配の符号が一緒なら次の通り:  
  $sign(\partial \xi /\partial w(i)) = sign(\partial \xi /\partial w(i-1))$
    * 重みの更新値 $\Delta$ には $\eta^+$をかけてください、なお  $\eta^+ > 1$です。  
    $\Delta(i) = \Delta(i) * \eta^+$
  * 勾配の逆向きに重みの更新値 $\Delta$ で重みを更新します。
  $w(i+1) = w(i) - sign(\partial \xi /\partial w(i))* \Delta(i)$
